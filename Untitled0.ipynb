```python
n0 = np.array([200,0,0,0], dtype=float)
L = np.array([[0,2,1,1],[0.5,0,0,0],[0,0.2,0,0],[0,0,0.1,0]])

total = [n0.sum()]
for i in range(200):
    n0 = L @ n0
    total.append(n0.sum())

print(n0)

time = range(len(total))
plt.plot(time, total)
plt.xlabel("Time step")
plt.ylabel("Total population")
plt.title("Total Population vs Time")
plt.grid(True)
# plt.yticks(range(0, int(max(total)) + 500, 500))
plt.show()
```

```python
n0 = np.array([200,0,0,0], dtype=float)
L = np.array([[0,2,1,1],[0.5,0,0,0],[0,0.2,0,0],[0,0,0.1,0]])

r21, r32, r43 = [], [], []

for i in range(200):
    r21.append(n0[1]/n0[0] if n0[0] != 0 else np.nan)
    r32.append(n0[2]/n0[1] if n0[1] != 0 else np.nan)
    r43.append(n0[3]/n0[2] if n0[2] != 0 else np.nan)
    n0 = L @ n0

plt.plot(r21, label='n2/n1')
plt.plot(r32, label='n3/n2')
plt.plot(r43, label='n4/n3')
plt.xlabel("Time step")
plt.ylabel("Ratio")
plt.legend()
plt.grid(True)
plt.show()
```

```python
n0 = np.array([200,0,0,0], dtype=float)
L = np.array([[0,2,1,1],[0.5,0,0,0],[0,0.2,0,0],[0,0,0.1,0]])

total = [n0.sum()]
r = []
for i in range(200):
    n0 = L @ n0
    total.append(n0.sum())
    r.append(total[-1]/total[-2])

print(n0)

plt.plot(r,"o",  label='N(t+1)/N(t)', markersize=2)
plt.xlabel("Time step")
plt.ylabel("Ratio")
plt.grid(True)
plt.legend()
plt.show()
```
